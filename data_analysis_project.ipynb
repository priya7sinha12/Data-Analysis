# Data Analysis of cars dataset

### importing libraries and loading dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Cars Data1.csv")
```

### exploring dataset

```python
data.head()
```

```python
data.shape
```

```python
data = data.drop_duplicates()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

### handling missing values

```python
data['Cylinders'].fillna(data['Cylinders'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

### data visualization

```python
data['Make'].value_counts()
```

```python
data.head(2)
```

```python
plt.figure(figsize=(6,6))
sns.histplot(data["Type"])
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(x="Type",data=data)
plt.title("Type Graph")
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(x="Origin",data=data)
plt.title("Origin Graph")
plt.show()
```

```python
# Visualize vehicle types
plt.figure(figsize=(8, 6))
data['Type'].value_counts().plot(kind='bar')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.title('Distribution of Vehicle Types')
plt.show()
```

```python
# Scatter plot for Horsepower vs. MPG_City
plt.figure(figsize=(8, 6))
plt.scatter(data['Horsepower'], data['MPG_City'])
plt.xlabel('Horsepower')
plt.ylabel('MPG (City)')
plt.title('Horsepower vs. MPG (City)')
plt.show()
```

```python
# Convert price columns to numeric values
data['MSRP'] = data['MSRP'].str.replace('$', '').str.replace(',', '').astype(float)
data['Invoice'] = data['Invoice'].str.replace('$', '').str.replace(',', '').astype(float)
# Grouping by Origin and calculating average MSRP and Invoice
grouped_data = data.groupby('Origin').agg({'MSRP': 'mean', 'Invoice': 'mean'})
print(grouped_data)
```

```python
# Box plot for MPG_City distribution by Origin
plt.figure(figsize=(8, 6))
sns.boxplot(x='Origin', y='MPG_City', data=data)
plt.xlabel('Origin')
plt.ylabel('MPG (City)')
plt.title('MPG (City) Distribution by Origin')
plt.show()
```

```python
# Create a new column for Fuel Efficiency (average of MPG_City and MPG_Highway)
data['Fuel_Efficiency'] = (data['MPG_City'] + data['MPG_Highway']) / 2

# Create a new column for Price Difference (difference between MSRP and Invoice)
data['Price_Difference'] = data['MSRP'] - data['Invoice']
```
